```python
import numpy as np
import pandas as pd
```

```python
#input
path_file = "./wiki_data/wiki-Vote.txt"
#output
path_df_nodes = "./wiki_data/df_nodes.csv"
path_df_edges = "./wiki_data/df_edges.csv"
path_df_nodes_subsample = "./wiki_data/df_nodes_subsample.csv"
path_df_edges_subsample = "./wiki_data/df_edges_subsample.csv"
```

```python
data_file = open(path_file,'r')
data = []
try: 
    for lines in data_file:
        data.append(lines.split())
finally:
    data_file.close()
```

```python
data = np.array(data)
df_data = pd.DataFrame(data,columns = ['node1','node2'])
```

```python
df_data
```

```python
df_nodes = pd.concat([df_data['node1'],df_data['node2']]).reset_index()
df_nodes.columns = ['in','node']
df_nodes = df_nodes.drop_duplicates(['node'],'first')['node'].reset_index()
df_nodes['Id'] = df_nodes['node']
df_nodes = df_nodes[['Id','node']]
df_nodes.columns = ['Id','Label']
df_nodes['Label'] = df_nodes['Label'].apply(lambda x: 'user ' + str(x))
df_nodes
```

```python
df_nodes_subsample = df_nodes.sample(n=2000)
```

```python
df_nodes_subsample
```

```python
df_nodes_subsample.to_csv(path_df_nodes_subsample)
```
